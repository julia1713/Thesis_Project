### APPLICATION OF VADER SENTIMENT ANALYZER  TO NEWS DATASET

```python
import pandas as pd
import matplotlib.pyplot as plt #plot the sentiment on a chart
import nltk 
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer #perform sentiment analysis 
```

```python
data=pd.read_csv('news_senti.csv')
```

```python
pd.set_option('display.max_colwidth',None)
data
```

```python
data=data.drop(columns='score and label')
```

```python
vader = SentimentIntensityAnalyzer() 

#compound: The normalized compound score which calculates the sum of all lexicon ratings and takes values from -1 to 1

# Iterate through the headlines and get the polarity scores using vader
f = lambda title: vader.polarity_scores(title)['compound']
data['compound'] = data['headline'].apply(f)

data
```

```python
def format_output(output_dict):
  
    polarity = "neutral"

    if output_dict['compound']>= 0.05:
        polarity = "positive"

    elif output_dict['compound']<= -0.05:
        polarity = "negative"

    return polarity

def predict_sentiment(text):
  
    output_dict =  vader.polarity_scores(text)
    return format_output(output_dict)

# Run the predictions
data["vader_prediction"] = data["headline"].apply(predict_sentiment)

# Show 5 random rows of the data
data.sample(5)
```

```python
data
```

```python
data['label'].value_counts()
```

```python
data['vader_prediction'].value_counts()
```

```python
data.rename(columns = {'label':'finbert_prediction'}, inplace = True)
```

```python
data=data.drop(columns='score')
```

```python
data=data.drop(columns='compound')
```

```python
data
```

```python
data['comparison'] = (data['finbert_prediction'] == data['vader_prediction'])
```

```python
data
```

```python
data['comparison'].value_counts()
```

```python
data.to_csv('news_all_senti.csv', index=False, encoding='utf-8-sig')
```

```python
#conclusion: finBERT is much more reliable!!!!!
```

### APPLICATION OF VADER SENTIMENT ANALYZER ALSO TO TWEETS DATASET

```python
data=pd.read_csv('tweets_senti.csv')
data=data.drop(columns='score and label')
data=data.drop(columns='Unnamed: 0')
```

```python
data
```

```python
vader = SentimentIntensityAnalyzer() 

#compound: The normalized compound score which calculates the sum of all lexicon ratings and takes values from -1 to 1

# Iterate through the headlines and get the polarity scores using vader
f = lambda title: vader.polarity_scores(title)['compound']
data['compound'] = data['text'].apply(f)

data
```

```python
def format_output(output_dict):
  
    polarity = "neutral"

    if output_dict['compound']>= 0.05:
        polarity = "positive"

    elif output_dict['compound']<= -0.05:
        polarity = "negative"

    return polarity

def predict_sentiment(text):
  
    output_dict =  vader.polarity_scores(text)
    return format_output(output_dict)

# Run the predictions
data["vader_prediction"] = data["text"].apply(predict_sentiment)

# Show 5 random rows of the data
data.sample(5)
```

```python
data.rename(columns = {'label':'finbert_prediction'}, inplace = True)
data=data.drop(columns='score')
data=data.drop(columns='compound')
```

```python
data
```

```python
data['comparison'] = (data['finbert_prediction'] == data['vader_prediction'])
```

```python
data['comparison'].value_counts()
```

```python
data.to_csv('tweets_all_senti.csv', index=False, encoding='utf-8-sig')
```
